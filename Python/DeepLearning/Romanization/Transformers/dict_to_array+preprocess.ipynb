```python
from hangul_utils import split_syllables, join_jamos
import pickle
import re
```

```python
with open("final_dict.pkl", "rb") as f:
    result_dict = pickle.load(f)
```

```python
# The data contains a lot of "길" which makes it predict unnecessary amount of gil in inference phase
# for key in result_dict.keys():
#     if "길" in key:
#         print(key, result_dict[key])
```

```python
# We will remove the gils
keys = list(result_dict.keys())
for key in keys:
    if re.search(r'길$', key):
        newkey = re.sub(r'길$', '', key)
        result_dict[newkey] = re.sub(r'-gil$', '', result_dict.pop(key))
#         print(result_dict[newkey])
```

```python
# The data contains a lot of "로" which makes it predict unnecessary amount of gil in inference phase
# for key in result_dict.keys():
#     if "로" in key:
#         print(key, result_dict[key])
```

```python
# We will remove the ros
keys = list(result_dict.keys())
for key in keys:
    if re.search(r'-ro$', result_dict[key]): # ro is a bit tricky as it has daero and ro. We will only remove the latter
        newkey = re.sub(r'로$', '', key)
        result_dict[newkey] = re.sub(r'-ro$', '', result_dict.pop(key))
        print(result_dict[newkey])
```

```python
# "고속도로" is translated as "Expressway"
```

```python
keys = list(result_dict.keys())
for key in keys:
    if re.search(r'Expressway$', result_dict[key]):
        newkey = re.sub(r'고속도로$', '', key)
        result_dict[newkey] = re.sub(r'Expressway$', '', result_dict.pop(key))
        print(result_dict[newkey])
```

```python
# Validate
for key in result_dict.keys():
    if "길" in key:
        print(key, result_dict[key])
```

```python
for key in result_dict.keys():
    if "로" in key:
        print(key, result_dict[key])
```

```python
for key in result_dict.keys():
    if "고속도로" in key:
        print(key, result_dict[key])
```

```python
# Split the keys and values
kor_token = []
eng_token = []
for key in result_dict.keys():
    kor_token.append(split_syllables(key))
    eng_token.append(result_dict[key].lower()) # reduce the need for using capital letters
```
